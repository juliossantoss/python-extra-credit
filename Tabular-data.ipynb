## Tabular data and graphling lesson
Tabular data is generally presented in rows and columns. You might have analyzed this type of data in a spreadsheet program before.

```python
import numpy
```

```python
help(numpy.genfromtxt)
```

```python
filepath = 'data/distance_data_headers.csv'

distances = numpy.genfromtxt(fname=filepath, delimiter=',', dtype='unicode' )
```

```python
print(distances)
```

```python
headers = distances[0]
```

```python
print(headers)
```

```python
data = distances[1:]
print(data)
```

```python
data = data.astype(float)
```

```python
print(data)
```

```python
# to access a specific element of array, array_name[row/column]
```

```python
print(data[2,1])
```

```python
point1 = data[0,1]
point2 = data[1,0] 
print(F'point2 is {point2}')
```

```python
small_data = data[0:10,0:3]
print(small_data)
```

```python
set1 = small_data[5,:]
set2 = small_data[:, 1:]
```

```python
print(set1)
```

```python
print(set2)
```

```python
#to take an average numpy.mean(data.set)
thr4_atp_data = data [:,1]
print(thr4_atp_data)
```

```python
average = numpy.mean(thr4_atp_data)
print(average)
```

```python
average = numpy.mean(data[:,1])
print(average)
```

```python
# range for loop does an operation a parrticular number of times
# range(start,end)
```

```python
#how do I find the number of columns?
#Using skills you already know, think about how you could find the number of columns.
num_columns = len(data[200,:])
print(num_columns)
```

```python
for i in range(1,num_columns):
    data_columns = data[:,i]
    average = numpy.mean(data_columns)
    print(F'{headers[i]} : {average:.4f}')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(data[:,1])
```

```python
plt.figure()
plt.plot(data[:,1], label=F'{headers[1]}')
plt.xlabel('simulation Frame')
plt.ylabel('distance (angstroms)')
plt.legend()

plt.savefig(F'{headers[1]}.png', dpi=300)
```

```python
plt.figure()
plt.plot(data[:,1], label=F'{headers[1]}', color='pink')
plt.plot(data[:,2], label=F'{headers[2]}', color='brown')

plt.xlabel('simulation Frame')
plt.ylabel('distance (angstroms)')
plt.legend()

plt.savefig(F'{headers[1]}.png', dpi=300)
```

```python
#using the example above, use a range for a loop to make a plot with all the columns of data
plt.figure()

for i in range(1, num_columns):
    plt.plot(data[:,i], label=F'{headers[i]}')
    
    plt.legend()
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstrom)')
```

```python
#think about how you would change the for loop to make every data set plot on its own graph.
colors = ['','red', 'green', 'pink', 'purple']

for i in range(1, num_columns):
    plt.figure()
    plt.plot(data[:,i], label=F'{headers[i]}', color=colors[i])
    plt.ylim(2,20)
    plt.legend()
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstrom)')
```
